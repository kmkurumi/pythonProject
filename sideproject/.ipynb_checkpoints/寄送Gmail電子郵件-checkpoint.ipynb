```python
import email
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.image import MIMEImage
import smtplib  #Simple Mail Transfer Protocol
from pathlib import Path
from string import Template
import sys  
```

```python
content=MIMEMultipart()  #建立MIMEMultipart物件
def send_email(mailtitle,email_account):
    content['subject']=mailtitle  #郵件標題
    content['from']='kmkurumi85@gmail.com'  #寄件者
    content['to']=email_account  #收件者
    content.attach(MIMEText('Dome python send email-1'))  #郵件內容
    content.attach(MIMEImage(Path('C:/Users/USER/Desktop/ina.jpg').read_bytes()))
    template = Template(Path("email_template.html").read_text())  #讀取檔案
    body = template.substitute({ "user": "Mike" })
    content.attach(MIMEText(body, "html"))  # HTML郵件內容
```

```python
def start():
    with smtplib.SMTP(host="smtp.gmail.com", port="587") as smtp:  # 設定SMTP伺服器
        try:
            smtp.set_debuglevel(1)
            #smtp.ehlo()  # 驗證SMTP伺服器
            smtp.starttls()  # 建立加密傳輸(TLS加密)
            smtp.ehlo()  # 驗證SMTP伺服器
            smtp.login('kmkurumi85@gmail.com', 'ojgnocynlfpydsel')  # 登入寄件者gmail，申請密碼
            smtp.send_message(content)  # 寄送郵件
            print("Complete!")
        except Exception as e:
            print("Error message: ", e)
```

```python
send_email('Python email_test','km19962@yahoo.com.tw')
start()
```
